```python
import matplotlib.pyplot as plt
import PIL
import os
import pandas as pd
from PIL import Image
```

```python
imgdir = '../crawl/save/'
files = os.listdir(imgdir)
```

```python
#for f in files:
```

```python
images = []
```

```python
df = pd.DataFrame(columns=['filename','x','y','format','mode'])
```

```python
# 포맷 png로 통일
for f in files:
    i = Image.open(imgdir+f)
    images.append(i)
    df = df.append(pd.DataFrame([[files[files.index(f)], i.width, i.height, i.format, i.mode]],
                               columns=['filename','x','y','format','mode']))
    os.rename(files[files.index(f)],'fire'+str(files.index(f)).zfill(5))
len(images)
```
